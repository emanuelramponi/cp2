#### 217. New Data: New York Citi SAT Scores

https://andybek.com/pandas-sat

```python
url_sat = 'https://andybek.com/pandas-sat'
sat = pd.read_csv(url_sat)
```

```python
sat.head()
```

```python
sat['Percent Tested']= sat['Percent Tested'].replace(regex='%',value='').astype(float)
```

```python
sat.info()
```

#### 218. Pivoting Data

```python
sat.set_index(['School Name', 'SAT Section'])
```

```python
sat.set_index(['School Name','SAT Section']).loc[:,['Score']]
```

```python
sat.set_index(['School Name','SAT Section']).loc[:,['Score']].unstack(level=1)
```

```python
# 3 steps in one line
```

```python
pivoted = sat.pivot(index='School Name',columns='SAT Section',values='Score')
```

```python
pivoted
```

#### 219. Undoing Pivots

```python
pivoted.stack().reset_index()
```

```python
# the melt() method
pivoted.melt()
# but we lose School Names because is in the index
```

```python
pivoted.reset_index().melt()
# the school names is considerer a measure variable but is not
```

```python
pivoted.reset_index().melt(id_vars='School Name')
```

```python
pivoted.reset_index().melt(id_vars='School Name',value_name='Score')
```

#### 221. The pivot_table()

```python
try:
    sat.pivot(index='Borough',columns='SAT Section',values='Score')
except Exception as e:
    print(f'Message Error: {e}')
```

```python
sat.pivot_table('Score','Borough','SAT Section')
```

```python
sat.pivot_table(values='Score',index='Borough',columns='SAT Section')
#by default de aggfunc  parameter is mean   
```

```python
sat.pivot_table(values='Score',index='Borough',columns='SAT Section',aggfunc='mean')
```

```python
sat.pivot_table(values='Score',index='Borough',columns='SAT Section',aggfunc='std')
```
