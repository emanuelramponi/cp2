#### 222. BONUS: The Problem With Average Percentage

```python
sat.pivot_table(values='Percent Tested',index='Borough',aggfunc='mean')
```

```python
# ====== ASIDE

# School A: 100 students, 30% took the SAT,
# School B: 1000 students, 60% took the SAT
```

```python
(.3 + .6) /2
```

```python
(30 + 600)/(100+1000)
```

That is why avergaes of porcenteages, equal weigh averages of percentages are very trickym potentially very mesleading

```python
# === END ASIDE
```

```python
sat.head()
```

```python
sat['SAT Takers'] = sat['Percent Tested'] * sat['Student Enrollment'] / 100
```

```python
sat.pivot_table(values='SAT Takers',index='Borough',aggfunc= 'mean')
```

```python
sat.pivot_table(values='Student Enrollment',index='Borough',aggfunc= 'mean')
```

```python
# these numbers are also misleading because boroughs have different sizes
```

```python
sat.groupby(['Borough'])[['Student Enrollment','SAT Takers']].apply(lambda sg: sg['SAT Takers'].mean() / sg['Student Enrollment'].mean())
```

```python
# So we see that the Bronx and Queens are the lowest the the boroughs with the lowest takers on average.

# And the numbers are for the Bronx.

# The number is very close to 60, but for Queens is almost 66, 67, and for Brooklyn it's 67.

# So if we go back to the beginning of this bonus lecture.

# Percentages that we were looking at here for Brooklyn 62 is very misleading.

# So first, Brooklyn is not the the second to lowest does not have the second to lowest percent tested.

# It's actually Queens.

# And second Brooklyn's levels are closer to 67% or 62.

# So we've seen how misleading averages of percentages are.

# And we've also seen how we can correct for that.
print(sat.pivot_table(values='Percent Tested',index='Borough',aggfunc='mean'))
print('==============================')
print(sat.groupby(['Borough'])[['Student Enrollment','SAT Takers']].apply(lambda sg: sg['SAT Takers'].mean() / sg['Student Enrollment'].mean()))
```
