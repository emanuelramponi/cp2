##### BONUS - Methods And Axes With fillna()

```python
# Este metodo completa el valor vacio con el valor del indice anterior, cuando esta con axis=0 lo hace de la columna, con axis=1 desde la fila
# ffill: propagate last valid observation forward to next valid.
# backfill / bfill: use next valid observation to fill gap.

# This filling methods are particularly useful when working with time series data because they allows us to plug the gaps in our series with a relatively
# reasonables values in our dataset.
soccer.fillna(method='ffill',axis=0).loc[[29,30,192,195]]
```

```python
soccer.fillna(method='ffill',axis=1).loc[[29,30,192,195]]
```

```python
soccer.fillna(method='bfill',axis=1).loc[[29,30,192,195]]
```

```python
soccer.fillna(method='bfill',axis=0).loc[[29,30,192,195]]
```

```python
def random_case(x):
    funcs = [x.str.swapcase,x.str.upper,x.str.lower,x.str.title]
    return choice(funcs)()
```

```python
for i in range(4):
    print(soccer.select_dtypes(include=object).transform(random_case).head(5))
```
