```python
from datetime import datetime
counter = 0
inflation = 1.02
def log_and_transfor(x):
    global counter
    counter += 1
    if counter % 100 == 0:
        print(f"It is {datetime.now()} and I just adjusted the {counter}th value.")
    
    return x * inflation
```

##### Skill Challenge

1. Create a standalone function that
    - Accepts a single parameter x
    - Returns the string 'relatively unkown' if x is less than 220
    - 'kind of popular' if x is between 220 and 600 (non-inclusive)
    - 'popular' if x is between 600 and 2000(non-inclusive)
    - 'super-popular' otherwise

2. Apply the function from the step above to the players page_views column. Use a method that supports vectorized operations
3. Add the output from the step above as a new column to the soccer dataframe. Name the column popularity
4. How many 'super-popular' players are there?

```python
#1 
def population(x):
    if x < 220:
        return 'relatively unkown'
    elif x < 600:
        return 'kind of popular'
    elif x < 2000:
        return 'popular'
    else:
        return 'super-popular'
```

```python
#2
soccer.page_views.apply(population)
#3
soccer['popularity'] = soccer.page_views.apply(population)
```

```python
#4 
soccer.loc[soccer.popularity.eq('super-popular')].shape[0]
soccer.popularity.value_counts()
soccer[soccer.popularity == 'super-popular'].name.size
```
