##### Comparators Operator Pandas Method
##### <     -->                   .lt()
##### <=    -->                   .le()
##### >        -->                .gt()
##### >=          -->             .ge()
##### ==             -->          .eq()

```python
print(soccer.age.le(25).equals(soccer.age <= 25 ))
print(soccer.market_value.le(40).equals(soccer.market_value <= 40))
soccer[soccer.age.eq(20)].shape[0]
```

##### ISIN()

```python
soccer.loc[soccer.position.isin(['LB','CB','RB'])].head()
```

##### BETWEEN


```python
soccer.loc[soccer.age.between(20,30,inclusive='neither')].sort_values(by='age',ascending=False).head(6)
```

```python
soccer.loc[(soccer.age.between(20,30,inclusive='neither') & (soccer.position.isin(['LB','CB','RB'])))].nlargest(4,'market_value')
```

##### conditionals: LB, 25 years old and market_value es 10m+

```python
soccer.loc[(soccer.position.eq('LB') & (soccer.age.le(25) & (soccer.market_value.ge(10))))]
```

##### conditions: LB, 25 de edad o mas joven, market_value 10 + y que no sea del Arsenal o Totthenham

```python
soccer[(soccer.position.eq('LB')) &
       (soccer.age.le(25) ) &
       (soccer.market_value.ge(10)) &
       ~(soccer.club.isin(['Arsenal','Tottenham']))
       ]
```

##### Skill challenge
1. The´re English(Nationality)
2. Their market value is more tha twice the average market value in the league(market_value) and
3. The either have more than 4000 views (page_views) or are a new signing (new_signing) but not both

```python
#1 
english = soccer.nationality.eq('England')

#2
valor = soccer.market_value.mean() * 2
valor_mas_del_doble = soccer.market_value.gt(valor)

#3
pg = (soccer.page_views.gt(4000)) | (soccer.new_signing.eq(1))



todo_junto = soccer.loc[english & valor_mas_del_doble & pg]
```
