##### 177. Cross Sections With xs()

```python
 # xs() is a type of label-based indexing
```

```python
tech.xs('2019-01-02')
```

```python
tech.loc['2019-01-02']
```

```python
tech.loc[(slice(None),'FB'),:].head()
```

```python
tech.xs('FB',level=1).head()
```

```python
tech.xs('FB',level=1,drop_level=False).head()
```

```python
# selecting from multiple level at once
tech.xs(('2019-01-02','FB'),level=(0,1))
```
