#### 188.  BONUS: Creating MultiLevel Columns Manually

LET´S BUILD ONE MANUALLY:

CRITERIA:
- 2-LEVEL MULTIINDEX COLUMN AXIS, CONTAINING:
    - TRADING VOLUME (HIGH AND LOW) AND TICKERS (MSFT AND AMZN)
    - 10 RANDOMN CLOSING PRICE OBSERVATIONS

```python
tech.reset_index(inplace= True)
```

```python
tech.head()
```

##### step 1: create multiindex

```python
cols = pd.MultiIndex.from_product([['low','high'],['MSFT','AMZN']], names = ['Volume Category', 'Ticker'])
```

```python
cols
```

##### step 2: prep the undelying data

```python
low = tech['Volume Categoy'] == 'low'
high = tech['Volume Categoy'] == 'high'
amzn = tech['Ticker'] == 'AMZN'
msft = tech['Ticker'] == 'MSFT'
```

```python
data = [
    tech[low & msft].close.sample(10).values,
    tech[low & amzn].close.sample(10).values,
    tech[high & msft].close.sample(10).values,
    tech[high & amzn].close.sample(10).values
]
```

```python
data = [ [*i] for i in zip(*data)]
data
```

```python
df = pd.DataFrame(data,columns=cols)
df
```

```python
df.columns.nlevels
```
