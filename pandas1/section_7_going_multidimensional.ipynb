#### 184. Multiple sort_index()

```python
try:
    tech.loc[(slice('2014-01-02','2014-04-02'),slice(None),slice('AAPL','FB')),'open':'low']
except Exception as e:
    print("Error: ",e)
```

```python
tech.sort_index(inplace=True)
```

```python
tech.loc[(slice('2014-01-02','2014-04-02'),slice(None),slice('AAPL','FB')),'open':'low']
```

Always consider sorting the index
Advantages:
- improves retrieval performance, which becomes significant
    - gor large dataframes, or
    - frequent retrieval
- enables slicing syntax
- overall a good practice when working with tabular data representations, including pandas, Excel, SQL, etc


##### sorting different levels in different orders

```python
tech.sort_index(level=(0,2), ascending=[False,True])
```
